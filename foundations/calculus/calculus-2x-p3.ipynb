# Chain Rule and Backpropagation 

```python
import numpy as np
```

```python
# f(g(x)) = sin(x²)
# g(x) = x²,     g'(x) = 2x
# f(u) = sin(u), f'(u) = cos(u)
# chain rule: cos(x²) * 2x

def numerical_derivative(f, x, tiny=1e-5):
    return (f(x + tiny) - f(x - tiny)) / (2 * tiny)
    
def h(x):
    return np.sin(x**2)

x = 1.5
numerical = numerical_derivative(h, x)
analytical = np.cos(x**2) * 2*x    # chain rule applied

print("Numerical: ", numerical)
print("Chain rule:", analytical)
print("Match:", np.allclose(numerical, analytical))
```

### Checking chain rule = backpropagation

```python
# A simple 2-layer computation:
# x → [multiply by w1] → z → [square it] → [multiply by w2] → loss
#
# loss = w2 * (w1 * x)²
# 
# We want: d(loss)/d(w1)  — how does loss change if we nudge w1?
# Chain rule: d(loss)/d(w1) = d(loss)/d(z) * d(z)/d(w1)

x  = 2.0
w1 = 0.5
w2 = 3.0

# forward pass — compute each step
z    = w1 * x          # first layer output
loss = w2 * z**2       # second layer output

print(f"z = {z}, loss = {loss}")

# backward pass — chain rule
d_loss_d_z  = w2 * 2 * z        # d(loss)/d(z) = 2 * w2 * z
d_z_d_w1    = x                  # d(z)/d(w1) = x
d_loss_d_w1 = d_loss_d_z * d_z_d_w1   # chain rule

print(f"d(loss)/d(w1) = {d_loss_d_w1}")

# verify numerically
eps = 1e-5
loss_plus  = w2 * ((w1 + eps) * x)**2
loss_minus = w2 * ((w1 - eps) * x)**2
numerical  = (loss_plus - loss_minus) / (2 * eps)
print(f"Numerical check:  {numerical}")
print(f"Match: {np.allclose(d_loss_d_w1, numerical)}")
```
